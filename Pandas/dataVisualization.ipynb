```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [4,5,1]
```

```python
plt.plot(x,y,color="red",linewidth='5')
plt.title('Info')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid('true',color="blue")
plt.show()
```

```python
# Use of style

from matplotlib import style
style.use('ggplot')

x=[5,8,10]
y=[12,16,6]
x2=[6,9,11]
y2=[6,15,7]

plt.plot(x,y,'g',label='First Line',linewidth=3)
plt.plot(x2,y2,'r',label='Second Line',linewidth=3)
plt.title('My Graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()

plt.show()
```
